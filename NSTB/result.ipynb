```python
import pandas as pd
```

```python
marks = {
    "obj":{
        "1": [9, 8, 10, 7, 10, 8, 4, 4, 5,5, 8],
        "2": [4, 3, 4, 3, 4, 6, 4, 3, 1, 6, 6],
        "3": [8, 7, 5, 7, 7, 7, 8, 8, 6, 8, 7],
        "4": [8, 7, 8, 5, 7, 9, 9, 9, 8, 7, 8 ],
        "5" : [4, 3, 3, 3, 4, 3, 4, 4, 4, 3, 3],
        "6": [5, 5, 4, 5, 4, 4, 4, 4, 4, 4, 4]
        },
    
    "subj": {
        
        "1": [7.5, 7, 7, 5.5, 6.5, 7.5, 7.5, 7.5, 8.5, 7, 6],
        "2": [5, 5, 5.5, 2.5, 4, 5, 5, 4, 5, 5, 5],
        "3": [7.5, 6, 6.5, 7, 7.5, 7.5, 3, 6, 5, 4.5, 5.5],
        "4": [8, 6, 5.5, 5, 6.5, 6.5, 7, 6, 2.5, 3.5, 3.5],
        "5" : [8, 6, 8, 6.5, 7.5, 7, 6, 6, 6.5, 6, 8],
        "6": [4.5, 3.5, 4.5, 3.5, 4, 4.5, 4, 4, 4, 3.5, 4]
        
    }
    
}
```

```python
import numpy as np
obj = np.array(list(marks["obj"].values())).T
subj = np.array(list(marks["subj"].values())).T
```

```python
obj.shape, subj.shape
```

```python
obj_wt =  np.array([10, 10, 8, 12, 5, 5])
obj_wt[:5]+=obj_wt[5]
```

```python
obj_wt
```

```python
subj[:, :5]+=subj[:, 5:6]
```

```python
subj
```

```python
obj[:, :5]+=obj[:, 5:6]
```

```python
obj
```

```python
normalized_obj = obj/obj_wt
```

```python
normalized_sub = subj/15
```

```python
result = (normalized_obj + normalized_sub) /2 
```

```python
r = result[:, :5] > 0.6 
```

```python
result_a = np.empty_like(result[:, :5], dtype='<U4')
```

```python
result_a[r] = "pass"
result_a[~r] = 'fail'
```

```python
result = pd.DataFrame(result_a, columns=[f"Unit-{i}" for i in range(1, 6)])
```

```python
result.to_csv('result.csv')
```
