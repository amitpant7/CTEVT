```python
import pandas as pd

# Load all sheets as a dictionary of DataFrames
all_dfs = pd.read_excel("reg_data_final.xlsx", sheet_name=None)
```

```python

from pathlib import Path


# Output directory
out_dir = Path("csv_outputs")
out_dir.mkdir(exist_ok=True)

processed_dfs = {}

for sheet_name, df in all_dfs.items():
    df = df.copy()

    # Extract column names from row 6
    column_names = df.iloc[6].values
    df.columns = column_names

    # Drop header/meta rows and reset index
    df = df.iloc[8:].reset_index(drop=True)

    # Save as CSV
    csv_path = out_dir / f"{sheet_name}.csv"
    df.to_csv(csv_path, index=False)

    # Keep reference if needed later
    processed_dfs[sheet_name] = df
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 180)
pd.set_option("display.max_colwidth", None)
```

```python
for sheet_name, df in processed_dfs.items():
    print(f"=== Sheet: {sheet_name} ===")
    try:
        from IPython.display import display
        display(df.describe())
    except Exception:
        print(df.describe())
    print()
```

```python
for sheet_name, df in processed_dfs.items():
    df.dropna(inplace=True, thresh=df.shape[0] / 2, axis=1)
    df['Quota'] = pd.to_numeric(df['Quota'], errors='coerce').abs()
    df['Applicant'] = pd.to_numeric(df['Applicant'], errors='coerce')
    
    
    while True:
        last = df.iloc[-1, 0]
        if last in ('Grand Total', 'Total'):
            print(True)
            df = df.iloc[:-1]
        else:
            break

    
    # merging for nan
    df.iloc[:, 1] = (
    df.iloc[:, 1]
      .ffill()
      .infer_objects(copy=False)
    )
    
    df.iloc[:, 2] = df.iloc[:, 2].ffill().infer_objects(copy=False)
    
    df.iloc[:, 0] = df.iloc[:, 0].ffill().infer_objects(copy=False)
    df = df.sort_values(by=df.columns[2])
    
    processed_dfs[sheet_name] = df
```

```python
for sheet_name, df in processed_dfs.items():
    csv_path = out_dir / f"{sheet_name}.csv"
    df.to_csv(csv_path, index=False)
```

```python
import pandas as pd

# Load all sheets as a dictionary of DataFrames
raw_dfs = pd.read_excel("diploma_year_stats.xlsx", sheet_name=None)
```

```python
df = raw_dfs['2081']
df = df.iloc[6:].dropna(how='all', axis=0).reset_index(drop=True)
```

```python

all_dfs = []
prev = 0
prev_name = ""
for row in df.iterrows():
    name = row[1].values[0]
    
    if type(name) != str:
        continue
    
    if "Program:" in name:
        name = name.replace("Program:", "").strip()
        curr = int(row[0])        
        if prev_name != "":
            section_df = df.iloc[prev+1:curr].reset_index(drop=True)
            col_names = ["S.N", "Institute Name", "District", "Approved"]   #"Quota"
            section_df.columns = col_names
            section_df["Program"] = prev_name
            all_dfs.append(section_df.iloc[2:-1].reset_index(drop=True))
        
        prev_name = name
        prev = curr
```

```python
df_2081 = pd.concat(all_dfs).reset_index(drop=True)
```
